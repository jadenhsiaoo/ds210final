```python
import csv

# Replace these with the names of your input CSV and output text files
input_csv = "musae_git_edges.csv"
output_txt = "github.txt"

# Set the column indices you want to extract (0-based)
id_1 = 0
id_2 = 1

# Read the CSV file and write the desired columns to the text file
with open(input_csv, 'r') as csv_file, open(output_txt, 'w') as txt_file:
    csv_reader = csv.reader(csv_file)

    for row in csv_reader:
        if len(row) > max(id_1, id_2):
            column1_value = row[id_1]
            column2_value = row[id_2]
            column1_value = row[id_1].replace(',', '')

                        # Write the column values to the text file, separated by a comma or any other separator
            txt_file.write(f"{column1_value}, {column2_value}\n")
```
